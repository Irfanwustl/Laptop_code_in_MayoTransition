```python
f1 = '/Users/irf3irf3/Desktop/offline_workspace/data/tissue_of_origin/Not Normalized/ArptRest_5_Samples_motifs_4.txt'
f2 = '/Users/irf3irf3/Desktop/offline_workspace/data/tissue_of_origin/Not Normalized/BWA_output_Arpit_motifs_4.txt'
f3 = '/Users/irf3irf3/Desktop/offline_workspace/data/tissue_of_origin/Not Normalized/Arpit_WGS_batch2_sorted_motifs_4.txt'

df1 = pd.read_csv(f1,sep ='\t',index_col = 0)
df2 = pd.read_csv(f2, sep = '\t', index_col = 0)
df3 = pd.read_csv(f3, sep = '\t', index_col = 0)

display(df1.head())
display (df2.head())
display (df3.head())
```
