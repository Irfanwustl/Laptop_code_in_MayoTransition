```python
# Function to process each group for imputation
def impute_group(group):
    group = group.copy()
    for column in group.columns:
        if group[column].notna().mean() >= 0.6:
            group[column] = group[column].fillna(group[column].mean())
    return group

# First, identify features that meet the 60% threshold in every group
columns_to_keep = promMeth_T.groupby('Target').apply(lambda g: g.notna().mean() >= 0.6)
columns_to_keep = columns_to_keep.all()  # Only keep columns that meet the criteria in all groups

# Filter the DataFrame to keep only these columns
promMeth_T_filtered = promMeth_T.loc[:, columns_to_keep]

# Now, impute missing values within each group for the retained columns
promMeth_T_cleaned = promMeth_T_filtered.groupby('Target').apply(impute_group)

# Reset the index without dropping the 'Target' column
promMeth_T_cleaned = promMeth_T_cleaned.reset_index(level=0, drop=False)

# Display the cleaned DataFrame
print(promMeth_T_cleaned.shape)
display(promMeth_T_cleaned.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_group_boxplot(df):
    """
    This function takes a DataFrame with a 'Target' column and real-valued features,
    and generates a box plot showing the distribution of all feature values grouped by the 'Target' column.

    Parameters:
    df (pd.DataFrame): The input DataFrame where 'Target' is the grouping variable and 
                       other columns are features with real values.

    Returns:
    None: The function will display the box plot.
    """
    
    # Melt the DataFrame to long format
    df_long = df.melt(id_vars='Target', var_name='Feature', value_name='Value')

    # Create a box plot for each Target group
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='Target', y='Value', data=df_long)

    # Set plot labels and title
    plt.xlabel('Group')
    plt.ylabel('Feature Values')
    plt.title('Box Plot of Feature Values Grouped by Target')

    # Show the plot
    plt.show()

# import matplotlib.pyplot as plt
# import seaborn as sns
# from scipy.stats import mannwhitneyu
# import itertools
# import pandas as pd

# def plot_group_boxplot(df):
#     """
#     This function takes a DataFrame with a 'Target' column and real-valued features,
#     and generates a box plot showing the distribution of all feature values grouped by the 'Target' column.
#     It also performs the Mann-Whitney U test for all pairs of groups in the 'Target' column for each feature,
#     printing the results.
    
#     Parameters:
#     df (pd.DataFrame): The input DataFrame where 'Target' is the grouping variable and 
#                        other columns are features with real values.

#     Returns:
#     None: The function will display the box plot and print MWU test results.
#     """
    
#     # Melt the DataFrame to long format
#     df_long = df.melt(id_vars='Target', var_name='Feature', value_name='Value')

#     # Create a box plot for each Target group
#     plt.figure(figsize=(12, 8))
#     sns.boxplot(x='Target', y='Value', data=df_long)

#     # Set plot labels and title
#     plt.xlabel('Group')
#     plt.ylabel('Feature Values')
#     plt.title('Box Plot of Feature Values Grouped by Target')

#     # Show the plot
#     plt.show()

#     # Perform MWU test for each feature
#     target_groups = df['Target'].unique()
    
#     for feature in df.columns:
#         if feature != 'Target':
#             print(f'\nMWU Test results for Feature: {feature}')
#             for group1, group2 in itertools.combinations(target_groups, 2):
#                 group1_data = df[df['Target'] == group1][feature]
#                 group2_data = df[df['Target'] == group2][feature]
                
#                 # Perform MWU test
#                 stat, p_value = mannwhitneyu(group1_data, group2_data, alternative='two-sided')
                
#                 # Print results
#                 print(f'{group1} vs {group2}: U={stat}, p-value={p_value}')

# Example usage
# Assuming 'df' is your DataFrame
#plot_group_boxplot(df)


# Example usage
plot_group_boxplot(promMeth_T_cleaned)
```

```python
import pandas as pd

def select_most_variable_features(df, num_features=1000):
    """
    Selects the most variable features from a DataFrame.

    Parameters:
    df (pd.DataFrame): The input DataFrame where 'Target' is the grouping variable
                       and other columns are features with real values.
    num_features (int): The number of most variable features to select.

    Returns:
    pd.DataFrame: A DataFrame containing the 'Target' column and the selected features.
    """
    
    # Separate the features and the Target column
    features = df.drop(columns=['Target'])
    target = df['Target']
    
    # Calculate the variance of each feature
    variances = features.var()

    # Select the top 'num_features' most variable features
    top_features = variances.nlargest(num_features).index

    # Create a new DataFrame with the most variable features
    df_top_features = df[top_features]
    
    # Add the 'Target' column back to the DataFrame
    df_top_features['Target'] = target

    return df_top_features

# Example usage
df_top_1000 = select_most_variable_features(promMeth_T_cleaned, num_features=1000)

# Now, plot the box plot for these selected features
plot_group_boxplot(df_top_1000)
```

```python
def subset_dataframe(dataframe, features_to_subset, target_column='Target'):
    """
    Subsets the given DataFrame to include only the specified features and the target column.
    
    Parameters:
    - dataframe: The DataFrame to subset.
    - features_to_subset: A list of feature names to keep in the subset.
    - target_column: The name of the target column to include in the subset (default is 'Target').
    
    Returns:
    - A subset of the original DataFrame containing only the specified features and the target column.
    """
    # Filter the list to include only features that are present in the DataFrame
    filtered_features = [feature for feature in features_to_subset if feature in dataframe.columns]

    # Ensure the target column is included
    filtered_features.append(target_column)

    # Subset the DataFrame
    df_subset = dataframe[filtered_features]

    return df_subset
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import seaborn as sns

def plot_pca(df):
    """
    Perform PCA on the given DataFrame and plot the first two principal components.
    The points are colored according to the 'Target' column.
    
    Parameters:
    - df: The input DataFrame with samples as the index, 'Target' as one column, and features as the other columns.
    """
    # Extract features and target
    features = df.drop(columns=['Target'])
    target = df['Target']
    
    # Standardize the features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)
    
    # Perform PCA
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(scaled_features)
    
    # Create a DataFrame with the PCA results
    pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
    pca_df = pd.concat([pca_df, target.reset_index(drop=True)], axis=1)
    
    # Plot the PCA results
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x='PC1', y='PC2', hue='Target', data=pca_df, palette='Set1')
    plt.title('PCA of Samples')
    plt.xlabel(f'PC1 - {pca.explained_variance_ratio_[0]*100:.2f}% Variance')
    plt.ylabel(f'PC2 - {pca.explained_variance_ratio_[1]*100:.2f}% Variance')
    plt.show()

# Example usage
# Assuming 'df' is your DataFrame
#plot_pca(df)
```

```python
def remove_rows_by_target(df, target_values_to_exclude):
    """
    Returns a subset of the DataFrame excluding rows where the 'Target' column matches any value in the provided list.
    
    Parameters:
    - df: The input DataFrame.
    - target_values_to_exclude: A list of values in the 'Target' column to exclude.
    
    Returns:
    - A subset of the DataFrame with the specified rows excluded.
    """
    # Filter out rows where the 'Target' column matches any value in target_values_to_exclude
    df_filtered = df[~df['Target'].isin(target_values_to_exclude)]
    
    return df_filtered
```

```python
from scipy.stats import mannwhitneyu
import itertools
import pandas as pd

def find_discriminatory_genes_with_medians(df):
    """
    This function performs the Mann-Whitney U test for all pairs of target groups 
    for each gene, including the median values for each group and the difference 
    between these medians. All results are returned sorted by p-value for significance.

    Parameters:
    - df (pd.DataFrame): The input DataFrame where 'Target' is the grouping variable and 
                         other columns are features (genes).

    Returns:
    - A dictionary where the keys are tuples representing each pair of target groups,
      and the values are DataFrames containing all discriminatory genes with their corresponding U statistics,
      p-values, median values for each group, and the difference between these medians, sorted by significance.
    """
    # Get unique target groups
    target_groups = df['Target'].unique()
    
    # Dictionary to store the discriminatory genes for each pair
    discriminatory_genes = {}
    
    # Iterate through each pair of target groups
    for group1, group2 in itertools.combinations(target_groups, 2):
        results = []
        
        # Iterate through each gene (feature) in the DataFrame
        for gene in df.columns:
            if gene != 'Target':
                group1_data = df[df['Target'] == group1][gene]
                group2_data = df[df['Target'] == group2][gene]
                
                # Perform MWU test
                stat, p_value = mannwhitneyu(group1_data, group2_data, alternative='two-sided')
                
                # Calculate the median values for each group
                median_group1 = group1_data.median()
                median_group2 = group2_data.median()
                
                # Calculate the difference between the median values
                median_diff = median_group1 - median_group2
                
                # Store the results
                results.append((gene, stat, p_value, median_group1, median_group2, median_diff))
        
        # Convert results to a DataFrame and sort by p-value (and then by U statistic)
        results_df = pd.DataFrame(results, columns=['Gene', 'U_statistic', 'p_value', f'{group1}_median', f'{group2}_median', f'{group1}-{group2}_median_diff'])
        results_df = results_df.sort_values(by=['p_value', 'U_statistic'], ascending=[True, False])
        
        # Store all results for the pair
        discriminatory_genes[(group1, group2)] = results_df
    
    return discriminatory_genes
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_volcano(df, group1, group2, p_value_threshold=0.05, median_diff_threshold=(-0.1, 0.1), annotation=False):
    """
    Creates a volcano plot for the given DataFrame that contains the results of MWU tests,
    and returns a subset of the DataFrame containing only the Significant & Large Effect points.

    Parameters:
    - df (pd.DataFrame): The DataFrame containing genes, U_statistic, p_value, median values, and median differences.
    - group1 (str): The name of the first group (used for labeling).
    - group2 (str): The name of the second group (used for labeling).
    - p_value_threshold (float): The threshold for highlighting significant p-values (default is 0.05).
    - median_diff_threshold (tuple): A tuple representing the range of median differences to be considered as having small effect (default is (-0.1, 0.1)).
    - annotation (bool): Whether to annotate significant & large effect points (default is False).

    Returns:
    - significant_large_effect_df (pd.DataFrame): A subset of the DataFrame containing only the Significant & Large Effect points.
    """

    # Calculate -log10(p_value) for plotting
    df['-log10(p_value)'] = -np.log10(df['p_value'])
    
    # Create a new figure for the volcano plot
    plt.figure(figsize=(10, 8))
    
    # Plot all points
    plt.scatter(df[f'{group1}-{group2}_median_diff'], df['-log10(p_value)'], color='grey', alpha=0.7)
    
    # Highlight significant points based on p_value_threshold
    significant = df['p_value'] < p_value_threshold
    plt.scatter(df.loc[significant, f'{group1}-{group2}_median_diff'], 
                df.loc[significant, '-log10(p_value)'], 
                color='red', alpha=0.7, label='Significant')
    
    # Highlight points with large median differences (outside the given range)
    large_effect = (df[f'{group1}-{group2}_median_diff'] < median_diff_threshold[0]) | \
                   (df[f'{group1}-{group2}_median_diff'] > median_diff_threshold[1])
    plt.scatter(df.loc[large_effect, f'{group1}-{group2}_median_diff'], 
                df.loc[large_effect, '-log10(p_value)'], 
                color='blue', alpha=0.7, label='Large Effect')
    
    # Highlight points that are both significant and have a large effect
    both = significant & large_effect
    plt.scatter(df.loc[both, f'{group1}-{group2}_median_diff'], 
                df.loc[both, '-log10(p_value)'], 
                color='green', alpha=0.7, label='Significant & Large Effect')
    
    # Annotate significant & large effect points if annotation is True
    if annotation:
        for i in df[both].index:
            plt.text(df.loc[i, f'{group1}-{group2}_median_diff'], 
                     df.loc[i, '-log10(p_value)'], 
                     df.loc[i, 'Gene'],
                     fontsize=8, ha='right')
    
    # Add labels and title
    plt.xlabel(f'{group1} - {group2} Median Difference')
    plt.ylabel('-log10(p_value)')
    plt.title(f'Volcano Plot: {group1} vs {group2}')
    
    # Add horizontal line for p-value threshold
    plt.axhline(-np.log10(p_value_threshold), color='black', linestyle='--')
    
    # Add vertical lines for median difference thresholds
    plt.axvline(median_diff_threshold[0], color='black', linestyle='--')
    plt.axvline(median_diff_threshold[1], color='black', linestyle='--')
    
    # Add a legend
    plt.legend()
    
    # Show the plot
    plt.show()
    
    # Return the subset of the DataFrame containing only the Significant & Large Effect points
    significant_large_effect_df = df[both]
    return significant_large_effect_df
```

```python
significant_large_effect_genes = plot_volcano(all_genes_between_group1_group2, 'Post_Sono', 'Pre_Sono', p_value_threshold=0.05, median_diff_threshold=(-0.1, 0.1), annotation=True)
```
