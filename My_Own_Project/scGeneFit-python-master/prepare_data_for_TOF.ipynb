```python
import pandas as pd
```

```python
infile = '../../../data/My_own_project/TCGA_ready_monod2_intersectedSortedMerged_std_filtered_cpg5_std0.1_sorted_Allmatrix'
tcga = pd.read_csv(infile,sep='\t')
```

```python
print(tcga.shape)
display(tcga.head())
```

```python
# Assuming GU list and additional substrings
GU = ["Bladder", "Kidney", "Prostate", "Testis"]
substrings = ['_Tumor', 'Metastatic', '_Normal', '_Cancer']

# Filter columns based on both GU substrings and the specified substrings
filtered_columns = [col for col in tcga.columns if any(gu in col for gu in GU) and any(sub in col for sub in substrings)]

# Create a copy of the DataFrame with the filtered columns and feature column
tcga_filtered = tcga[['feature'] + filtered_columns].copy()

print(tcga_filtered.shape)

# Rename columns for kidney samples and general GU Normal format
def rename_gu_sample(col):
    # Check if column is "Kidney" and matches any substring
    if "Kidney" in col:
        for sub in substrings:
            if sub in col:
                split_part = col.split(sub, 1)[1]
                col = f"Kidney{sub}{split_part}"
    
    # Rename any column with *_Normal to GuNormal format
    if "_Normal" in col:
        split_part = col.split("_Normal", 1)[1]
        col = f"GuNormal_Normal{split_part}"
    
    return col

# Apply renaming function to columns
tcga_filtered.columns = [rename_gu_sample(col) for col in tcga_filtered.columns]

# Display the result
display(tcga_filtered.head())
```

```python
# Count columns containing each GU category and GuNormal
gu_counts = {gu: sum(gu in col for col in tcga_filtered.columns) for gu in GU}
gu_normal_count = sum("GuNormal_Normal" in col for col in tcga_filtered.columns)

# Display the counts
print("Counts of each GU category:")
print(gu_counts)
print("\nCount of GuNormal_Normal columns:", gu_normal_count)

```

```python
# Define the input file name (assuming you have already set infile)
# infile = "your_input_file_name"  # Uncomment and set this if infile is not already defined

# Construct the output file name without an extension
output_file = infile + "_TOF"

# Save the DataFrame as a tab-separated file without the index
tcga_filtered.to_csv(output_file, sep='\t', index=False)

print(f"File saved as {output_file}")
```
